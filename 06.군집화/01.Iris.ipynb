## K-Means Algorithm - Iris

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')       # warnings 무시해라
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
iris_std = StandardScaler().fit_transform(iris.data)

pca = PCA(2)
iris_pca = pca.fit_transform(iris_std)
```

#### 1. 군집화(Clustering)

```python
from sklearn.cluster import KMeans
KMeans().get_params()
```

```python
kmeans = KMeans(n_clusters=3, random_state=2023, n_init='auto')  # n_clusters : 군집의 갯수, n_init:  초기 중심위치 시도 횟수,  실행과정 보고 싶으면 verbose=1 옵션추가
kmeans.fit(iris_std)        # 비지도 학습이므로 y값을 제공하지 않음
```

```python
# 학습 결과
kmeans.labels_
```

```python
# numbering은 임의로 부여   0=49개, 1=55개, 2=46개  (* 0,1,2는 타겟이 지정되지 않아서 무슨 꽃인줄 모른다!!)
np.unique(kmeans.labels_, return_counts=True)
```

#### 2. 시각화

```python
df = pd.DataFrame(iris_pca, columns= ['PC1', 'PC2'])
df['target'] = iris.target
df['cluster'] = kmeans.labels_
df.head(54).tail()
```

```python
# 데이터 분포
df.groupby(['target', 'cluster'])[['PC1']].count()
```

```python
markers = ['^', 's', 'o']
plt.figure(figsize=(6,4))
for i, marker in enumerate(markers):
    x_data = df[df.cluster == i]['PC1']
    y_data = df[df.cluster == i]['PC2']
    plt.scatter(x_data, y_data, marker=marker, label='cluster'+str(i))
    plt.scatter(x_data.mean(), y_data.mean(), marker='*', c='red', s=100)       #c=color, s=size
plt.legend()
plt.xlabel('PC1'), plt.ylabel('PC2')
plt.title('3 Clustering result')
plt.show()
```

#### 3. Original Data vs. Clustering result 

```python
plt.figure(figsize=(12,4))
for k, column in enumerate(['target', 'cluster']):
    plt.subplot(1, 2, k+1)
    for i, marker in enumerate(markers):
        x_data = df[df[column] == i]['PC1']
        y_data = df[df[column] == i]['PC2']
        if column == 'target':
            plt.scatter(x_data, y_data, marker=marker, label=iris.target_names[i])
        else:
            plt.scatter(x_data, y_data, marker=marker, label='cluster'+str(i))
            plt.scatter(x_data.mean(), y_data.mean(), marker='*', c='red', s=100)
    plt.legend()
    plt.xlabel('PC1'), plt.ylabel('PC2')
    title = 'Origianl Data' if column=='target' else '3 Clustering result'
    plt.title(title)
plt.show()
```
